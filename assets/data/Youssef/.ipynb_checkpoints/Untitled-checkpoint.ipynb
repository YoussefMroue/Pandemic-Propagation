```python
import os
import pandas as pd
import numpy as np
import operator
import datetime
```

```python
h1file = os.path.join("H1N1-Data.csv")
rawh1 = pd.read_csv(h1file,header= 0,encoding= 'ISO-8859-1')
```

```python
rawh1case = rawh1
rawh1case.drop(['Deaths'], axis = 1)
rawh1case = rawh1case[['Update Time','Country','Cases']]
rawh1case = rawh1case.reindex(index = rawh1case.index[::-1])
rawh1case = rawh1case.reset_index()
rawh1case = rawh1case.drop(['index'],axis = 1)
rawh1case = rawh1case[rawh1case.Country != "Grand Total"]
rawh1case
```

```python
rawh1case.loc[rawh1case.Country == "Costa Rica*", "Country"] = "Costa Rica"
```

```python
dates =(rawh1case['Update Time'])
dates = dates.drop_duplicates()
dates = dates.transpose()
dates = dates.values.tolist()
countries = pd.DataFrame(rawh1case['Country'])
countries = countries.drop_duplicates(subset = 'Country')
countries = countries.reset_index()
countries = countries.drop(['index'], axis = 1)
template = countries
template.head()
```

```python
for i in range(len(dates)):
    template[dates[i]] = ""
template = template.set_index('Country')
```

```python
for index, row in rawh1case.iterrows():
     template.loc[row[1],row[0]] = row[2]
```

```python
template.head()
```

```python
dates = template.columns
dates = pd.to_datetime(dates)
dates = dates.strftime('%m-%d-%Y')
template.columns = dates
dates = [datetime.datetime.strptime(date, '%m-%d-%Y').date() for date in dates]
for i in range(5):
    for dif in range(len(dates)):
        if (dates[dif+1]-dates[dif]).days>1:
            addme = dates[dif] + datetime.timedelta(days=1)
            dates.insert(dif+1, addme)
dates = [date_obj.strftime('%m-%d-%Y') for date_obj in dates]
```

```python
for i in range(len(dates)):
    if dates[i] != template.columns[i]:
        template.insert(loc = i,column = dates[i],value = template[template.columns[i-1]].tolist())
```

```python
template.head()
```

```python
days = []
for x in range(len(template.columns)):
    days.append('Day '+str(x))
template.columns = days
```
